```
import numpy as np
import scipy as sp 
import pylab as pl
from scipy.optimize import leastsq #import least square function
```

```
n=9 #degree of polynomial
```

```
def real_f(x):
  return np.sin(2*np.pi*x) #real function f(x)=sin(2pi x)

def fit_f(p,x):
  f=np.poly1d(p) #polynomial function
  return f(x)  

def residual_f(p,y,x):
  ret = fit_f(p,x)-y #aproximation error
  return ret
```

```
x=np.linspace(0,1,9) #Randomly points x
x_points=np.linspace(0,1,1000) 

y0=real_f(x) #real function output 
y1=[np.random.normal(0,0.1)+y for y in y0]
p_init=np.random.randn(n) #Randomly initialize polynomial parameters 
plsq=leastsq(residual_f,p_init,args=(y1,x))

print('Fitting Parameters:', plsq[0]) #Output fitting parameters 
pl.plot(x_points,real_f(x_points),label='real')
pl.plot(x_points,fit_f(plsq[0],x_points), label='fitted curve')
pl.plot(x,y1,'go',label='with noise')
pl.legend()
pl.grid(True)
pl.show()
```
