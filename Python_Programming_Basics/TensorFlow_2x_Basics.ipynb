### Tensor Slicing and indexing 

  - Tensor slicing methods include: 
      - [start: end] extracts a data slice from the start position to the end position of the tensor. 
      - [start: end: step] or [::step] extracts a data slice at an invertval of step from the start position to the end position of the tensor. 
      - [::-1] slices data from the last element. 


```
#Create a 4-D tensor. 
#The tensor contains 5 images. The size of each image is 100x100x3
img_tensor=tf.random.normal([5,100,100,3])
print(img_tensor)
```

```
#Extract the first image
img_tensor[0,:,:,:]
```

```
#Extract one slice at an interval of two images
img_tensor[::2,...]
```

```
#Slice data from the least element
img_tensor[::-1]
```

### Indexing

```
#The basic format of an index is a[][][]

#Obtain the pixel in the position [20,40] in the second channel of the firt image
img_tensor[0][19][39][1]
```

```
# Dimension Reshaping 
a=tf.constant([[1,2,3],[4,5,6]])
print(a)
tf.reshape(a,(1,6))
```

```
# Transpose 
print(tf.transpose(a))
```

## AutoGraph of TensorFlow 2.x

When used to comment out a function, the decorator tf.function can be called like any other function. tf.function will be compiled into a graph, so that it can rum more efficiently on a GPU or TPU. In this case, the function becomes an operation in TensorFlow. The function can be directly called out a return value. However, the function is executed in graph mode and intermediate variable cannot be directcly viewed. 

```
@tf.function 
def simple_nn_layer(w,x,b):
  print(b)
  return tf.nn.relu(tf.matmul(w,x) +b) #layer output

w=tf.random.uniform((3,3)) #Random matrix 3x3 
x=tf.random.uniform((3,3)) #Random matrix 3x3
b=tf.constant(0.5,dtype='float32')

#call the function
simple_nn_layer(w,x,b)

'''
According to the output result, the value of b in the function cannot be
viewed directly, but the return value can be viewed using .numpy()
'''
```
