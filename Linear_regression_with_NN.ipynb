This experiment mainly uses keras library and the simplest data to reproduce how a linear regression algorithm iterates and fits the existing data with a feedfoward with a single hidden layer.

```
#import the necessary modules 
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
```

```
#define data, and change list to array 

x=[3,21,22,34,54,34,55,67,89,99]
x=np.array(x)

y=[1,10,14,34,44,36,22,67,79,90]
y=np.array(y)

#Show the effect of a scatter plot
plt.scatter(x,y)
```

```
#Feedforward network
model = keras.Sequential()

inputs = keras.Input(shape=(1,), name = 'x')# Define inputs
xc=layers.Dense(10, activation='linear', name = 'layer1')(inputs)
output=layers.Dense(1, activation='linear', name = 'output')(xc)
model = keras.Model(inputs=inputs, outputs=output)    
model.summary()
```

```
#optimizer=Adam(lr=1e-4,loss="mean_squared_error")
model.compile(loss='mean_squared_error', optimizer=Adam(lr=1e-3))
```

```
history = model.fit(x, y, epochs=500)
```

```
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.plot(x, model.predict(x), color='r')
```
